````python
# pip install rpy2
# call R files
import rpy2.robjects as robjects
r_source = robjects.r['source']
r_source("multivariate_bernoulli_clean.R")


# call R functions
r_gen = robjects.r['generate_synthetic_sequences']

# r_returns = r_gen(1000, "Files/Real_intubation_alignment_annoymous.csv")
r_returns = r_gen(1000, "Files_synthetic/alignment.csv")

print(len(r_returns))
````

````python
import numpy as np
import matplotlib.pyplot as plt

generated = r_returns[0] # generated data
data = r_returns[1]  # real data
data = np.matrix(data)
generated = np.matrix(generated)
np.savetxt('Files_synthetic/Bernoulli_generated_1000.csv', generated, delimiter=',')
print(type(data))
print(data)
print(generated.shape)
print(np.sum(generated[:,34]))

len_data = np.sum(data,axis = 1)
len_generated = np.sum(generated, axis = 1)
print(np.mean(len_generated), "+/-", np.std(len_generated))
print(np.mean(len_data), "+/-", np.std(len_data))

len_generated = np.squeeze(np.asarray(len_generated))
len_data = np.squeeze(np.asarray(len_data))

plt.hist(x = [len_generated, len_data], bins = 24, normed = True, alpha = 0.5, histtype='stepfilled', label=['Generated', 'real']) # normalized
plt.legend()
plt.show()

plt.hist(x = [len_generated, len_data], bins = 24, normed = True, alpha = 0.5, histtype='bar', label=['Generated', 'real']) # normalized
plt.legend()
plt.show()
````
